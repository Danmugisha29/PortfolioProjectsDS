# New football dataframe

```python
football_player_stats_new_df = pd.DataFrame()
```

```python
football_player_stats_new_df['Player'] = football_player_stats_new['Player']
football_player_stats_new_df['Nation'] = football_player_stats_new['Nation']
football_player_stats_new_df['Pos'] = football_player_stats_new['Pos']
football_player_stats_new_df['Comp'] = football_player_stats_new['Comp']
football_player_stats_new_df['Squad'] = football_player_stats_new['Squad']
football_player_stats_new_df['Age'] = football_player_stats_new['Age']
football_player_stats_new_df['MP'] = football_player_stats_new['MP']
football_player_stats_new_df['G/90'] = football_player_stats_new['Goals']
football_player_stats_new_df['G/Sh'] = football_player_stats_new['G/Sh']
football_player_stats_new_df['PKGoals'] = ((football_player_stats_new['ShoPK'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['PKAttempted'] = ((football_player_stats_new['PKatt'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['Goals'] = ((football_player_stats_new['Goals'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['Pass'] = ((football_player_stats_new['PasTotAtt'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['PassCompleted'] = ((football_player_stats_new['PasTotCmp'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['PassComp%'] = ((football_player_stats_new_df['PassCompleted'] / football_player_stats_new_df['Pass']) * 100).round(2)
football_player_stats_new_df['Assist'] = ((football_player_stats_new['Assists'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['Cross'] = ((football_player_stats_new['PasCrs'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['CrossCompleted'] = ((football_player_stats_new['CrsPA'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['CrossComp%'] = ((football_player_stats_new_df['CrossCompleted'] / football_player_stats_new_df['Cross']) * 100).round(2)
football_player_stats_new_df['Tackle_Won'] = ((football_player_stats_new['TklWon'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['SucDribble'] = ((football_player_stats_new['DriSucc'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['Dribble'] = ((football_player_stats_new['DriAtt'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['DribbleComp%'] = ((football_player_stats_new_df['SucDribble'] / football_player_stats_new_df['Dribble']) * 100).round(2)
football_player_stats_new_df['YCards'] = ((football_player_stats_new['CrdY'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['RCards'] = ((football_player_stats_new['CrdR'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['Fls'] = ((football_player_stats_new['Fls'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['Fld'] = ((football_player_stats_new['Fld'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['OGoals'] = ((football_player_stats_new['OG'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['AerWon'] = ((football_player_stats_new['AerWon'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['AerLost'] = ((football_player_stats_new['AerLost'] * football_player_stats_new['Min']) / 90).round(0).astype(int)
football_player_stats_new_df['AerWon%'] = ((football_player_stats_new['AerWon'] / (football_player_stats_new['AerWon'] + football_player_stats_new['AerLost'])) * 100).round(2)
```
