### Applying functions to rows
the_adder = lambda x: x + 210101

The map method will pass each element of a Series (remember a series is basically a column of a DataFrame)
 to the lambda function. The apply method will pass an entire row to the lambda 
In short this means that if you want to perform calculations using multiple columns of data, 
use the apply method, otherwise use the map method.

```python
# A genre_count column is added and is filled with elements
# then Pandas library allows us to pass each element of the Series in the `netflix["listed_in"]` to the lambda function as `x`. 
# and then perform the logic with the code after the colon.
netflix_dataset["genre_count"] = netflix_dataset["listed_in"].map(lambda x: len(x.split(",")))
netflix_dataset["genre_count"]
netflix_dataset
```

 ### Count the number of cast members
 
The apply method and lambda function to help perform operations on the column.

```python
netflix_dataset["cast"]

netflix_dataset["nonsensical_columns"] = netflix_dataset.apply(lambda x: len(x["cast"].split(",")) +
 len(x["listed_in"].split(",")), axis = 1)

netflix_dataset
```

### GroupBy, Aggregation, and Sorting Values

To perform aggregations on your data, the groupby function can be used to do those kinds of operations on the data.

```python
netflix_dataset.groupby("country")["show_id"].count()
netflix_dataset

netflix_dataset.groupby("country")["show_id"].count().reset_index()

netflix_movies_by_country = netflix_dataset.groupby("country")["show_id"].count().reset_index().sort_values(by="show_id", ascending=False)
netflix_movies_by_country
netflix_dataset
```

### Joins and Unions
In python, the pandas library allows us to help join different datasets together for better visualizations and anaylsis.

```python
# Here the second netflix dataset is read into the program
netflix_second_dataset = pd.read_csv('netflix_titles_second.csv')
netflix_second_dataset

# Then a new dataset that is the combination of both datasets is made with 
# the help of the pd.concat function.

new_dataset = pd.concat([netflix_dataset, netflix_second_dataset])
new_dataset
```

### Left Joins

The netflix dataset is combined using a left join with the `netflix_movies_by_country` data that was made earlier.

```python
netflix_merged = pd.merge(left = new_dataset, right = netflix_movies_by_country, how ="inner",
 left_on = ["country"], right_on = ["country"])

netflix_merged
```

#### Pivot table analysis

Oftentimes you'll want to pivot or "melt" data. Pivoting data takes it from the "long" format that we are used to and puts it in a "wide" format that might be easier to read.

```python
pivot_table = netflix_dataset.pivot_table(
    index="country", 
columns="type", 
values= "title", 
aggfunc='count',
 fill_value=0).reset_index()
pivot_table
```
